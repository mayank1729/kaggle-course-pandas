Maps

A map is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later. Maps are what handle this work, making them extremely important for getting your work done!

There are two mapping methods that you will use often.

map() is the first, and slightly simpler one. For example, suppose that we wanted to remean the scores the wines received to 0. We can do this as follows:

````python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p-review_points_mean)
````

The function you pass to map() should expect a single value from the Series (a point value, in the above example), and return a transformed version of that value. map() returns a new Series where all the values have been transformed by your function.

apply() is the equivalent method if we want to transform a whole DataFrame by calling a custom method on each row.

````python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row
````

````python
reviews.apply(remean_points,axis='columns')
````

### Understanding `axis` in pandas `apply()`

The `axis` parameter tells pandas **which direction to move across the DataFrame when applying a function**.

---

### Key idea

| axis | Direction pandas moves | Function applied to |
|-----|------------------------|---------------------|
| `axis=0` or `"index"` | ↓ down the rows | each **column** |
| `axis=1` or `"columns"` | → across the columns | each **row** |

---

### Example

```python
reviews.apply(remean_points, axis='columns')
```

What happens here:

- pandas **moves across the columns**
- collects the values from each column in that row
- passes the **entire row** to the function

So the function is applied **row-wise**.

Example DataFrame:

| country | points | price |
|--------|--------|-------|
| Italy | 90 | 25 |
| France | 88 | 30 |

`remean_points` will receive:

```
Row 1 → [Italy, 90, 25]
Row 2 → [France, 88, 30]
```

---

### Visual memory trick

```
axis = 0  ↓  (down rows)
operate on columns

axis = 1  →  (across columns)
operate on rows
```

---

### Another way to remember

If you write:

```python
axis='columns'
```

pandas **moves across columns**, which naturally processes **rows**.

---

### Quick rule

**`axis=1` → combine/use columns → operate on rows**

If we had called reviews.apply() with axis='index', then instead of passing a function to transform each row, we would need to give a function to transform each column.

Note that map() and apply() return new, transformed Series and DataFrames, respectively. They don't modify the original data they're called on. If we look at the first row of reviews, we can see that it still has its original points value.

Pandas provides many common mapping operations as built-ins. For example, here's a faster way of remeaning our points column:



````python
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
````

In this code we are performing an operation between a lot of values on the left-hand side (everything in the Series) and a single value on the right-hand side (the mean value). Pandas looks at this expression and figures out that we must mean to subtract that mean value from every value in the dataset.

## Why `dropna()` did not remove the `NaN` values from `reviews.country`

`dropna()` is working correctly, but not in the way it might appear at first.  
This happens because of **how pandas aligns indices when assigning a Series back to a DataFrame column**.

### What `dropna()` actually does

When running:

```python
reviews.country = reviews.country.dropna()
```

`reviews.country.dropna()` returns a **Series with the NaN rows removed**, but it **keeps the original index values**.

Example:

```
Original Series

index  country
0      US
1      NaN
2      Canada
3      France
```

After `dropna()`:

```
index  country
0      US
2      Canada
3      France
```

Notice that **index 1 is missing**, but the index numbers themselves are not reset.

### What happens during assignment

When assigning this Series back:

```python
reviews.country = reviews.country.dropna()
```

pandas **aligns values by index**:

```
reviews DataFrame

index  country
0      US
1      NaN
2      Canada
3      France
```

Since row **1 still exists in the DataFrame**, pandas fills it with `NaN` again.

Therefore, it may look like `dropna()` did not work, but in reality the DataFrame rows were never removed.

---

## Correct ways to remove the null values

### Option 1 (Recommended): Drop rows from the DataFrame

```python
reviews = reviews.dropna(subset=['country'])
```

This removes all rows where `country` is `NaN`.

---

### Option 2: Handle `NaN` safely in the function

```python
import pandas as pd

def ratings(row):
    if row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

    if pd.notnull(row.country) and row.country.lower() == "canada":
        return 3
```

---

### Option 3: Use pandas string operations

Convert to pandas' nullable string type:

```python
reviews.country = reviews.country.astype("string")
```

Or use the `.str` accessor:

```python
reviews.country.str.lower()
```

`.str` methods automatically **ignore `NaN` values**, preventing errors like:

```
AttributeError: 'float' object has no attribute 'lower'
```
