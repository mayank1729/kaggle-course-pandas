````python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row
````

````python
reviews.apply(remean_points,axis='columns')
````

### Understanding `axis` in pandas `apply()`

The `axis` parameter tells pandas **which direction to move across the DataFrame when applying a function**.

---

### Key idea

| axis | Direction pandas moves | Function applied to |
|-----|------------------------|---------------------|
| `axis=0` or `"index"` | ↓ down the rows | each **column** |
| `axis=1` or `"columns"` | → across the columns | each **row** |

---

### Example

```python
reviews.apply(remean_points, axis='columns')
```

What happens here:

- pandas **moves across the columns**
- collects the values from each column in that row
- passes the **entire row** to the function

So the function is applied **row-wise**.

Example DataFrame:

| country | points | price |
|--------|--------|-------|
| Italy | 90 | 25 |
| France | 88 | 30 |

`remean_points` will receive:

```
Row 1 → [Italy, 90, 25]
Row 2 → [France, 88, 30]
```

---

### Visual memory trick

```
axis = 0  ↓  (down rows)
operate on columns

axis = 1  →  (across columns)
operate on rows
```

---

### Another way to remember

If you write:

```python
axis='columns'
```

pandas **moves across columns**, which naturally processes **rows**.

---

### Quick rule

**`axis=1` → combine/use columns → operate on rows**
